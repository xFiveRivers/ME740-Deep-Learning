## Softmax Function

The Softmax function is a mathematical function commonly used in machine learning for multi-class classification problems. It is used to convert a vector of real numbers into a probability distribution. The output of the Softmax function assigns probabilities to each class, ensuring that the probabilities sum up to 1.

**Mathematical Formula:**

The Softmax function for class \(i\) is defined as:
$$
\text{Softmax}(x_i) = \frac{e^{x_i}}{\sum_{j=1}^{N} e^{x_j}}
$$
Where:

- $(\text{Softmax}(x)_i)$ is the probability of class \(i\).
- $(x_i)$ is the input value for class \(i\).
- $(N)$ is the total number of classes.
- $(e)$ is Euler's number, the base of the natural logarithm.

**How It Works:**

The Softmax function exponentiates each element of the input vector \(x\), making them all positive, and then normalizes them by dividing each element by the sum of all exponentiated values. This normalization ensures that the resulting values represent valid probabilities, with the sum equal to 1.




```python
import numpy as np

def softmax(x):
    # Compute the exponential of each element in the input array
    exp_x = np.exp(x)
    
    # Calculate the sum of the exponential values
    sum_exp_x = np.sum(exp_x)
    
    # Calculate the softmax probabilities
    softmax_probs = exp_x / sum_exp_x
    
    return softmax_probs

# Example input, where x is a vector of real numbers
x = np.array([2.0, 1.0, 0.1])

# Applying the softmax function to the input
softmax_result = softmax(x)

print("Input vector:", x)
print("Softmax probabilities:", softmax_result)
print("Sum of softmax probabilities:", np.sum(softmax_result))
```

## Softmax using PyTorch

```python
import torch
import torch.nn.functional as F

# Step 1: Create a tensor with input values (logits)
logits = torch.tensor([2.0, 1.0, 0.1])

# Step 2: Apply the Softmax function
softmax_probs = F.softmax(logits, dim=0)

# Step 3: Display the input logits and Softmax probabilities
print("Input Logits:", logits)
print("Softmax Probabilities:", softmax_probs)
```

## Entropy

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the entropy function
def entropy(p):
    return -p * np.log2(p) - (1 - p) * np.log2(1 - p)

# Generate a range of probabilities from 0 to 1 (excluding 0 and 1)
p_values = np.linspace(0.01, 0.99, 100)

# Calculate entropy for each probability
entropy_values = entropy(p_values)

# Create a plot with a black background
with plt.style.context('dark_background'):
    plt.figure(figsize=(8, 6))
    plt.plot(p_values, entropy_values, label='Entropy', color='lime')
    
    # Customize tick color
    plt.xticks(color='red')
    plt.yticks(color='red')
    
    # Customize label color
    plt.xlabel('Probability (p)', color='red')
    plt.ylabel('Entropy', color='red')
    
    # Customize title color
    plt.title('Entropy Function', color='red')
    
    # Add a grid
    plt.grid(True, color='gray', linestyle='--', alpha=0.6)
    
    # Customize legend color
    plt.legend(loc='upper right', fontsize='large', facecolor='black', edgecolor='red', labelcolor='red')
    
    # Show the plot
    plt.show()

```

```python
import numpy as np

# Define a probability distribution (as an array)
probabilities = np.array([0.2, 0.3, 0.1, 0.4])

# Calculate entropy
def entropy(p):
    return -np.sum(p * np.log2(p))

result_entropy = entropy(probabilities)

print("Probability Distribution:", probabilities)
print("Entropy:", result_entropy)
```

## Cross Entropy

```python
import numpy as np

# True labels
true_labels = np.array([0, 1])

# Predicted probabilities for class 1 (Dog)
predicted_probs = np.array([0.2,0.8])

# Calculate cross-entropy for each data point
cross_entropy_per_point = - (true_labels * np.log(predicted_probs))

# Calculate average cross-entropy
average_cross_entropy = np.sum(cross_entropy_per_point)

print("Cross-Entropy per Point:", cross_entropy_per_point)
print("Average Cross-Entropy:", average_cross_entropy)
```

```python
import torch
import torch.nn.functional as F

# True labels (as PyTorch tensor)
true_labels = torch.tensor([0], dtype=torch.float32)

# Predicted probabilities (as PyTorch tensor)
predicted_probs = torch.tensor([0.2], dtype=torch.float32)

# Calculate binary cross-entropy using PyTorch's loss function
cross_entropy = F.binary_cross_entropy(predicted_probs, true_labels)

print("Binary Cross-Entropy:", cross_entropy.item())
```

## Mean and Variance

```python
import numpy as np

# Sample data (replace this with your data)
data = np.array([3, 5, 7, 8, 10])

# Calculate the biased variance (population variance)
biased_variance = np.var(data)

# Calculate the unbiased variance (sample variance)
unbiased_variance = np.var(data, ddof=1)

print("Biased Variance:", biased_variance)
print("Unbiased Variance:", unbiased_variance)
```

```python
# Sample data (replace this with your data)
data = [3, 5, 7, 8, 10]

# Calculate the mean manually
mean = sum(data) / len(data)

# Calculate the biased variance manually
biased_variance = sum((x - mean)**2 for x in data) / len(data)

# Calculate the unbiased variance manually
unbiased_variance = sum((x - mean)**2 for x in data) / (len(data) - 1)

print("Mean:", mean)
print("Biased Variance:", biased_variance)
print("Unbiased Variance:", unbiased_variance)
```

## t_test 

```python
import numpy as np
from scipy import stats

# Sample data for two groups (Group A and Group B)
group_a = np.array([85, 88, 92, 78, 90, 82, 88, 86, 75, 80])
group_b = np.array([78, 80, 85, 80, 88, 92, 76, 89, 82, 85])

# Perform the independent t-test
t_statistic, p_value = stats.ttest_ind(group_a, group_b)

# Set the significance level (alpha)
alpha = 0.05

# Print the results
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

# Determine whether to reject the null hypothesis
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between the groups.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the groups.")
```

## Derivative

```python
import sympy as sp
from IPython.display import display, Math

# Define the symbolic variables
x, u = sp.symbols('x u')

# Define the composite function f(g(x))
g_x = 3 * x**2 - 2
f_u = u**3 + 1

# Compute the derivatives
g_prime_x = sp.diff(g_x, x)
f_prime_u = sp.diff(f_u, u)

# Apply the chain rule
chain_rule_result = f_prime_u.subs(u, g_x) * g_prime_x

# Display the explanation and result
display(Math("f(g(x)) = u^3 + 1"))
display(Math("g(x) = 3x^2 - 2"))
display(Math("f'(u) = 3u^2"))
display(Math("g'(x) = 6x"))
display(Math("f'(g(x)) = f'(u) * g'(x)"))
display(Math("f'(g(x)) = 3u^2 * 6x = 18x(u^2)"))
display(Math("f'(g(x)) = 18x(3x^2 - 2)^2"))

# Simplify the result
simplified_result = sp.simplify(chain_rule_result)
display(Math("f'(g(x)) = 18x(3x^2 - 2)^2 = " + sp.latex(simplified_result)))
```

```python
# Define the symbolic variables
x = sp.symbols('x')

# Define the functions u(x) and v(x)
u_x = 2 * x**2
v_x = sp.sin(x)

# Compute the derivatives
u_prime_x = sp.diff(u_x, x)
v_prime_x = sp.diff(v_x, x)

# Apply the product rule
product_rule_result = u_prime_x * v_x + u_x * v_prime_x

# Display the explanation and result
display(Math("u(x) = 2x^2"))
display(Math("v(x) = \\sin(x)"))
display(Math("u'(x) = 4x"))
display(Math("v'(x) = \\cos(x)"))
display(Math("w(x) = u(x) \cdot v(x)"))
display(Math("w'(x) = u'(x) \cdot v(x) + u(x) \cdot v'(x)"))
display(Math("w'(x) = (4x) \cdot (\\sin(x)) + (2x^2) \cdot (\\cos(x))"))

# Simplify the result
simplified_result = sp.simplify(product_rule_result)
display(Math("w'(x) = (4x) \cdot (\\sin(x)) + (2x^2) \cdot (\\cos(x)) = " + sp.latex(simplified_result)))
```
